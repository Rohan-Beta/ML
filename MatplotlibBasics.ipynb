```python
from matplotlib import pyplot as plt
```

```python
plt.plot([1 , 2 , 3] , [2 , 5 , 7])
plt.show() # first point 1,2 second 2,5 third 3,7
```

```python
plt.plot([1 , 2 , 3] , [2 , 5 , 7] , label = "graph line")
plt.title("Sample Graph")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
from matplotlib import style

style.use("ggplot")

x = [1 , 5 , 4]
y = [23 , 45 , 89]

x2 = [3 , 7 , 3]
y2 = [5 , 8 , 9]

plt.plot(x , y , label = "first" , linewidth = 3)
plt.plot(x2 , y2 , label = "second")
plt.legend()

plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x = [1 , 5 , 4]
y = [23 , 45 , 89]

x2 = [3 , 7 , 3]
y2 = [5 , 8 , 9]

plt.hist(x , label = "first" , linewidth = 3)
plt.plot(x2 , y2 , label = "second" , color = "blue")
plt.legend()
plt.show()
```

histogram example

```python
import numpy as np
```

```python
np.random.seed(19680801)
```

```python
mu , sigma = 100 , 15

x = mu + sigma * np.random.randn(10000)
```

```python
plt.hist(x , 50 , density = True , color = "green" , alpha = 0.75)

plt.xlabel("Smarts")
plt.ylabel("Probability")
plt.title("Histogram Example")
plt.text(60 , .025 , r"$\mu = 100 , \ \sigma = 15$")
plt.axis([40 , 160 , 0 , 0.03])
plt.grid(True)
plt.show()
```

merged dataframe or csv

```python
df4 = pd.DataFrame([[1 , 2 , 3] , [2 , 7 , 3] , [6 , 3 , 9]] , columns = ["a" , "b" , "c"])
df4
```

```python
df5 = pd.DataFrame([[11 , 7 , 312] , [22 , 2 , 35] , [63 , 3 , 92]] , columns = ["x" , "y" , "z"])
df5
```

```python
df6 = pd.merge(df4 , df5 , left_on= "b" , right_on = "y")
```
